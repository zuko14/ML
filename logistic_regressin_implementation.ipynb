```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_classification
```

```python
# creting the dataset
x,y=make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=15)
```

```python
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
```

```python
logistic.fit(x_train,y_train)
```

```python
y_pred=logistic.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
```

                                          # HYPERPARAMETER TUNING AND CROSS VALIDATION

     # grid search used for the best selection of the parameter for model better accuracy

```python
model=LogisticRegression()
penalty=['l1', 'l2', 'elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver=['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver)
```

```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
```

```python
# grid search

from sklearn.model_selection import GridSearchCV

grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
grid
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_score_
```

```python
param_grid1= [
    {'solver': ['liblinear'], 'penalty': ['l1', 'l2']},
    {'solver': ['lbfgs', 'newton-cg', 'sag'], 'penalty': ['l2']},
    {'solver': ['saga'], 'penalty': ['l1', 'l2', 'elasticnet'], 'l1_ratio': [0.5]}
]
```

```python
grid1=GridSearchCV(estimator=model,param_grid=param_grid1,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
grid1.fit(x_train,y_train)
```

```python
y_pred=grid1.predict(x_test)
```

```python
grid1.best_score_
```

```python
grid1.best_params_
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```

  # RANDOMIZED SEARCH CV

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
model=LogisticRegression()
randomcv=RandomizedSearchCV(estimator=model,param_distributions=params,cv=5,scoring='accuracy')
```

```python
randomcv.fit(x_train,y_train)
```

```python
randomcv.best_score_
```

```python
randomcv.best_params_
```

```python
y_pred=randomcv.predict(x_test)
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```

## Logistic Regression For Multiclass Classification Problem

```python
## create the dataset
x, y = make_classification(n_samples=1000, n_features=10,n_informative=3, n_classes=3, random_state=15)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression(multi_class='ovr')
logistic.fit(x_train,y_train)
y_pred=logistic.predict(x_test)
```

```python
y_pred
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```
